```python
import pandas as pd
import numpy as np
import scipy
from scipy import stats
from numpy import sqrt
```

```python
prot= pd.read_excel(r"C:\Users\Client\My Drive\PhD\Programing\R\Datasets\Cerebrospinal\Cereb No NA.xlsx", 
                    header=0, index_col=False)
prot 
```

```python
#df['average'] = df.mean(numeric_only=True, axis=1)
prot['Mean Control'] = prot.loc[:, prot.columns.str.startswith('Control ')].mean(axis=1)
prot['Mean ALS'] = prot.loc[:, prot.columns.str.startswith('ALS ')].mean(axis=1)
prot['Mean Mim'] = prot.loc[:, prot.columns.str.startswith('Mim ')].mean(axis=1)
prot['Mean PD'] = prot.loc[:, prot.columns.str.startswith('PD ')].mean(axis=1)
prot['Mean PLS'] = prot.loc[:, prot.columns.str.startswith('PLS ')].mean(axis=1)
prot
```

```python
#Standard diviation
prot['SD Control'] = prot.loc[:, prot.columns.str.startswith('Control ')].std(axis=1)
prot['SD ALS'] = prot.loc[:, prot.columns.str.startswith('ALS ')].std(axis=1)
prot['SD Mim'] = prot.loc[:, prot.columns.str.startswith('Mim ')].std(axis=1)
prot['SD PD'] = prot.loc[:, prot.columns.str.startswith('PD ')].std(axis=1)
prot['SD PLS'] = prot.loc[:, prot.columns.str.startswith('PLS ')].std(axis=1)
prot
```

```python
#Fold Change
prot["FC ALS-C"]= prot["Mean ALS"]/prot["Mean Control"]
prot["FC Mim-C"]= prot["Mean Mim"]/prot["Mean Control"]
prot["FC PD-C"]= prot["Mean PD"]/prot["Mean Control"]
prot["FC PLS-C"]= prot["Mean PLS"]/prot["Mean Control"]
prot
```

```python
#Log 2 FC
prot["Log2FC ALS-C"]=np.log2(prot["FC ALS-C"])
prot["Log2FC Mim-C"]=np.log2(prot["FC Mim-C"])
prot["Log2FC PD-C"]=np.log2(prot["FC PD-C"])
prot["Log2FC PLS-C"]=np.log2(prot["FC PLS-C"])
prot
```

```python
# counting the lenght of each group
n1 = len(prot.columns[prot.columns.str.startswith('Control ')])
n2 = len(prot.columns[prot.columns.str.startswith('ALS ')])
n3 = len(prot.columns[prot.columns.str.startswith('Mim ')])
n4 = len(prot.columns[prot.columns.str.startswith('PD ')])
n5 = len(prot.columns[prot.columns.str.startswith('PLS ')])
```

```python
#Calculating standard error
prot['SE Control'] = prot["SD Control"]/sqrt(n1)
prot['SE ALS'] = prot["SD ALS"]/sqrt(n2)
prot['SE Mim'] = prot["SD Mim"]/sqrt(n3)
prot['SE PD'] = prot["SD PD"]/sqrt(n4)
prot['SE PLS'] = prot["SD PLS"]/sqrt(n5)
prot
```

```python
# standard error on the difference between the samples
#sed = sqrt(se1**2.0 + se2**2.0)
prot['SED ALS-Control'] = sqrt(prot["SE Control"]**2 + prot['SE ALS']**2)
prot['SED Mim-Control'] = sqrt(prot["SE Control"]**2 + prot['SE Mim']**2)
prot['SED PD-Control'] = sqrt(prot["SE Control"]**2 + prot['SE PD']**2)
prot['SED PLS-Control'] = sqrt(prot["SE Control"]**2 + prot['SE PLS']**2)
prot
```

```python
# calculate the t statistic
#t_stat = (mean1 - mean2) / sed
prot["Tstat ALS-Control"] = (prot["Mean ALS"] - prot["Mean Control"])/prot['SED ALS-Control']
prot["Tstat Mim-Control"] = (prot["Mean Mim"] - prot["Mean Control"])/prot['SED Mim-Control']
prot["Tstat PD-Control"] = (prot["Mean PD"] - prot["Mean Control"])/prot['SED PD-Control']
prot["Tstat PLS-Control"] = (prot["Mean PLS"] - prot["Mean Control"])/prot['SED PLS-Control']
prot
```

```python
# calculate the P value, students T-test
prot["P Value ALS"] = (1 - stats.t.cdf(abs(prot["Tstat ALS-Control"]), n2-1)) * 2
prot["P Value Mim"] = (1 - stats.t.cdf(abs(prot["Tstat Mim-Control"]), n3-1)) * 2
prot["P Value PD"] = (1 - stats.t.cdf(abs(prot["Tstat PD-Control"]), n4-1)) * 2
prot["P Value PLS"] = (1 - stats.t.cdf(abs(prot["Tstat PLS-Control"]), n5-1)) * 2
prot
```

```python
#sort the P value for P adj value
sort_prot = prot.sort_values(by="P Value ALS", ascending=True, kind="mergesort")
sort_prot
```
