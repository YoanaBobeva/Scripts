```python
dat['SD ALL'] = dat.std(axis=1)
dat
```

```python
# df['average'] = df.mean(numeric_only=True, axis=1)
dat['Mean Control'] = dat.loc[:, dat.columns.str.startswith('Control ')].mean(axis=1)
dat['Mean ALS'] = dat.loc[:, dat.columns.str.startswith('ALS ')].mean(axis=1)
dat['Mean ALS-FTD'] = dat.loc[:, dat.columns.str.startswith('ALS-FTD ')].mean(axis=1)
dat['Mean FTD'] = dat.loc[:, dat.columns.str.startswith('FTD ')].mean(axis=1)
dat
```

```python
dat['SD Control'] = dat.loc[:, dat.columns.str.startswith('Control ')].std(axis=1)
dat['SD ALS'] = dat.loc[:, dat.columns.str.startswith('ALS ')].std(axis=1)
dat['SD ALS-FTD'] = dat.loc[:, dat.columns.str.startswith('ALS-FTD ')].std(axis=1)
dat['SD FTD'] = dat.loc[:, dat.columns.str.startswith('FTD ')].std(axis=1)

dat
```

```python
import numpy as np
from matplotlib import pyplot
```

```python
#Fold Change
dat["FC ALS-C"]= dat["Mean ALS"]/dat["Mean Control"]
dat["FC ALS-FTD-C"]= dat["Mean ALS-FTD"]/dat["Mean Control"]
dat["FC FTD-C"]= dat["Mean FTD"]/dat["Mean Control"]
dat
```

```python
#Log 2 FC
dat["Log2FC ALS-C"]=np.log2(dat["FC ALS-C"])
dat["Log2FC FTD-C"]=np.log2(dat["FC FTD-C"])
dat["Log2FC ALS-FTD-C"]=np.log2(dat["FC ALS-FTD-C"])
dat
```

```python
# counting the lenght of each group
n1 = len(dat.columns[dat.columns.str.startswith('Control ')])
n2 = len(dat.columns[dat.columns.str.startswith('ALS ')])
n3 = len(dat.columns[dat.columns.str.startswith('ALS-FTD ')])
n4 = len(dat.columns[dat.columns.str.startswith('FTD ')])
n1,n2,n3,n4
```

```python
from numpy import sqrt
```

```python
#Calculating standard error

dat['SE Control'] = dat["SD Control"]/sqrt(n1)
dat['SE ALS'] = dat["SD ALS"]/sqrt(n2)
dat['SE ALS-FTD'] = dat["SD ALS-FTD"]/sqrt(n3)
dat['SE FTD'] = dat["SD FTD"]/sqrt(n4)
dat
```

```python
# standard error on the difference between the samples
#sed = sqrt(se1**2.0 + se2**2.0)
dat['SED ALS-Control'] = sqrt(dat["SE Control"]**2 + dat['SE ALS']**2)
dat['SED ALS-FTD-Control'] = sqrt(dat["SE Control"]**2 + dat['SE ALS-FTD']**2)
dat['SED FTD-Control'] = sqrt(dat["SE Control"]**2 + dat['SE FTD']**2)
dat
```

```python
# calculate the t statistic
#t_stat = (mean1 - mean2) / sed
dat["Tstat ALS-Control"] = (dat["Mean ALS"] - dat["Mean Control"])/dat['SED ALS-Control']
dat["Tstat ALS-FTD-Control"] = (dat["Mean ALS-FTD"] - dat["Mean Control"])/dat['SED ALS-FTD-Control']
dat["Tstat FTD-Control"] = (dat["Mean FTD"] - dat["Mean Control"])/dat['SED FTD-Control']
dat
```

```python
import scipy
from scipy import stats
```

```python
dat["P Value ALS"] = (1 - stats.t.cdf(abs(dat["Tstat ALS-Control"]), n2-1)) * 2
dat["P Value ALS-FTD"] = (1 - stats.t.cdf(abs(dat["Tstat ALS-FTD-Control"]), n3-1)) * 2
dat["P Value FTD"] = (1 - stats.t.cdf(abs(dat["Tstat FTD-Control"]), n4-1)) * 2
dat
```

```python
sort_datA ["P adj ALS"] =(sort_datA["P Value ALS"]* len(sort_datA)/ sort_datA["Rank ALS P values"])
sort_datA
```
